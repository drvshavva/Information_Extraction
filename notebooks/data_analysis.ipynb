# create vocabulary db

```python
def get_list_from_dict(index: int, _dict: dict):
    """
    This method returns list from given index dict values
    """
    values_list = []
    for key in _dict[6]:
        for value in _dict[index][key]:
            values_list.append(value)
    return values_list
```

```python
# create fiil table
fiiler = {'fiiller': list(fiiller_dicts[0].keys())}
create_db_from_dict(fiiler, "fiiller")
```

```python
# create sıfatlar table
isimler_6 = get_list_from_dict(6, isimler_dicts) # sıfatları
fiiller_m_2 = get_list_from_dict(-2, fiiller_dicts) # nasıl yapılır ?
isimler_10 = get_list_from_dict(10, isimler_dicts) # renk
isimler_2 = get_list_from_dict(2, isimler_dicts) # şekil

sıfatlar = {'sıfatlar': list(set(isimler_2 + isimler_10 + fiiller_m_2 + isimler_6))}
create_db_from_dict(sıfatlar, "sıfatlar")
```

```python
# hammadde
isimler_0 = get_list_from_dict(0, isimler_dicts) # hammeddesi
isimler_4 = get_list_from_dict(4, isimler_dicts) # içinde neler bulunur ?

hammadde = {'hammedde': list(set(isimler_0 + isimler_4))}
create_db_from_dict(hammadde, "hammaddesi")
```

```python
# yer
isimler_1 = get_list_from_dict(1, isimler_dicts) # nerede bulunur ?
fiiller_3 = get_list_from_dict(3, fiiller_dicts) # nerede yapılır ?

yer = {'yer': list(set(isimler_1 + fiiller_3))}
create_db_from_dict(yer, "yer")
```

```python
# kim
isimler_8 = get_list_from_dict(8, isimler_dicts) # kim kullanır ?
fiiller_6 = get_list_from_dict(6, fiiller_dicts) # kim/ne yapar ?

kim = {'kim': list(set(isimler_8 + fiiller_6))}
create_db_from_dict(kim, "kim")
```

```python
# ne oldu
fiil_8 = get_list_from_dict(8, fiiller_dicts) # yapınca ne olur ?

ne_olur = list(set(fiil_8))
create_db_from_dict(ne_olur, "ne_olur")
```

```python
# ne zaman
fiil_10 = get_list_from_dict(10, fiiller_dicts) # ne olunca yapılır ?

ne_zaman = list(set(fiil_10))
create_db_from_dict(ne_zaman, "ne_zaman")
```
